```python
def simple_chatbot():
    print("Chatbot: Hello! I am a simple rule-based bot. (Type 'bye' to quit)")

    while True:
        user_input = input("You: ").lower().strip()

        if user_input == "hello":
            print("Chatbot: Hi!")
        elif user_input == "how are you":
            print("Chatbot: I'm fine, thanks!")
        elif user_input == "bye":
            print("Chatbot: Goodbye!")
            break
        else:
            print("Chatbot: I'm sorry, I don't understand that. Try 'hello', 'how are you', or 'bye'.")

# Run the chatbot
simple_chatbot()
```

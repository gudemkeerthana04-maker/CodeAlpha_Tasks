## Simple Stock Tracker

This notebook will help you calculate your total investment based on manually defined stock prices and user-entered quantities.

```python
# Define hardcoded stock prices
stock_prices = {
    "AAPL": 180.00,
    "TSLA": 250.00,
    "GOOG": 150.00,
    "AMZN": 175.00,
    "MSFT": 420.00
}

print("Available stocks and their prices:")
for stock, price in stock_prices.items():
    print(f"- {stock}: ${price:.2f}")
```

Now, let's get your stock quantities. You will be prompted to enter the quantity for each stock. If you don't own a stock, you can enter `0`.

```python
stock_holdings = {}
for stock_symbol in stock_prices.keys():
    while True:
        try:
            quantity = int(input(f"Enter quantity for {stock_symbol}: "))
            if quantity >= 0:
                stock_holdings[stock_symbol] = quantity
                break
            else:
                print("Quantity cannot be negative. Please enter a non-negative number.")
        except ValueError:
            print("Invalid input. Please enter a whole number.")

print("\nYour stock holdings:")
for stock, quantity in stock_holdings.items():
    if quantity > 0:
        print(f"- {stock}: {quantity} shares")
```

Next, we will calculate your total investment based on the current stock prices and your entered quantities.

```python
total_investment = 0.0
stock_investment_details = []

for stock_symbol, quantity in stock_holdings.items():
    if quantity > 0:
        price = stock_prices[stock_symbol]
        investment_value = quantity * price
        total_investment += investment_value
        stock_investment_details.append(f"{stock_symbol}: {quantity} shares @ ${price:.2f} = ${investment_value:.2f}")

print("\nInvestment Breakdown:")
for detail in stock_investment_details:
    print(detail)

print(f"\nTotal Investment Value: ${total_investment:.2f}")
```
